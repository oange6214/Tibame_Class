# 載入資料

```python
import pandas as pd

train_df = pd.read_csv('../train.csv', encoding='utf-8')
train_df
```

```python
test_df = pd.read_csv('../test.csv', encoding='utf-8')
test_df
```

# 資料欄位型態區分: 
    1. 數值 (大小)
        Age、SibSp、Parch、Fare
    2. 類別 (幾種)
        pclass、Name (mid)、Sex、Embarked
    3. 其他
        Ticket、Cabin(缺值太多)

# 補缺失值
    數值 (最可能值):
        中位數來補
    類別 (最常出現):
        最常出現的來補

```python
train_df.isna().sum()
```

```python
test_df.isna().sum()
```

```python
train_df.median()
```

# 補中位數
    Plass 不要計算

```python
med = train_df.median().drop(['Pclass'])
train_df = train_df.fillna(med)
train_df.isna().sum()
```

# 測試資料只用 訓練資料計算的 median 去補中位數

```python
test_df = test_df.fillna(med)
test_df.isna().sum()
```

# 補最常出現

```python
most = train_df['Embarked'].value_counts().idxmax()
train_df['Embarked'] = train_df['Embarked'].fillna(most)
train_df.isna().sum()
```

```python
test_df['Embarked'] = test_df['Embarked'].fillna(most)
test_df.isna().sum()
```

    # 做 one-hot encoding
    Pclass 可以不做 (有大小關係)
    Name (mid) 必須做
    Sex 可不做 (只有兩個值)
    Embarked 必須做

# 處理 name

```python
# import re

# s = "Braund, Mr. Owen Harris"
# pat = '(\w+),\s*(\w+)\.(.+)'
# re.search(pat, s).group(2)
```

```python
s = "Braund, Mr. Owen Harris"
s = s.split(',')[-1].split('.')[0]
s.strip()
```

```python
def name_cut(s):
    s = s.split(',')[-1].split('.')[0]
    return s.strip()

train_df['Name'].apply(name_cut).value_counts()
```

```python
def name_cut(s):
    reserved = ['Mr', 'Mrs', 'Miss', 'Master']
    s = s.split(',')[-1].split('.')[0]
    s = s.strip()
    if s in reserved:
        return s
    else:
        return None
    
train_df['Name'] = train_df['Name'].apply(name_cut)
train_df
```

```python
test_df['Name'] = test_df['Name'].apply(name_cut)
test_df
```

```python
train = pd.get_dummies(train_df,
                         columns=['Name', 'Sex', 'Embarked'])
predict = pd.get_dummies(test_df,
                           columns=['Name', 'Sex', 'Embarked'])
train
```

```python
x_train = train.drop(['PassengerId', 'Survived', 'Ticket', 'Cabin'], axis=1)
y_train = train['Survived']
x_predict = predict.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)

predict_id = predict['PassengerId']
```

# 交叉驗證

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier()
params = {
    'n_estimators': range(20, 35),
    'max_depth': range(6, 10)
}
grid = GridSearchCV(clf, params, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
print('最佳參數: ', grid.best_params_)
print('最佳分數 (10 次分數): ', grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(n_estimators=22, max_depth=7)
scores = cross_val_score(clf,
                         x_train,
                         y_train,
                         cv=10,
                         n_jobs=4)

print('十次分數: ', scores)
print('平均分數: ', sum(scores) / len(scores))
```

# 訓練

```python
clf = RandomForestClassifier(n_estimators=22, max_depth=7)
clf.fit(x_train, y_train)
```

```python
pre = clf.predict(x_predict)
result = pd.DataFrame({
    "PassengerId": predict_id,
    "Survived": pre
})

result.to_csv("titanic_rf.csv",
              encoding='utf-8',
              index=False)
result
```

# 特徵重要比例

```python
imp = clf.feature_importances_

d = sorted(zip(x_train.columns, imp), 
                   key=lambda x: x[1],
                   reverse=True)

pd.DataFrame({
    "Name": [n for n, im in d],
    "Importance": [im for n, im in d]
})
```

# 題外話
    組合方式: 
        平行 (Bagging)
            三個臭皮匠 > 一個諸葛亮的概念。(利用隨機放棄一點資料，達到較佳的好處)
        序列 (Boosting): Adaboost、 XGBoost... 
           加權處理，無法平行處理，大大影響效率。
           
        廣納演算法不一定有效，重點在重資料挑選。
        
    1. 特徵處理很重要
    2. 機率的差異不大，用甚麼模型都無所謂
    
    KNN 沒有 feature_important 所以無法確認哪個特徵重要

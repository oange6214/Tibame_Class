# 資料載入

```python
import pandas as pd

train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
test_df
```

# 資料處理

## 訓練、測試資料集合併處理
    1. 先將兩個表格規格調整一致
    2. 合併後移除非必要欄位

```python
train_df_drop = train_df.drop(["SalePrice"], axis=1)
datas = pd.concat([train_df_drop, test_df], axis=0)
datas = datas.drop(["Id"], axis=1)
datas
```

## 資料缺失檢視

```python
na = datas.isna().sum()
na[na != 0].sort_values(ascending=False)
```

## drop 掉缺失值過多的欄位

```python
datas = datas.drop(na[na > 1000].index, axis=1)
datas
```

## 填缺失值(數字)

```python
med = datas.median().drop(["MSSubClass"])
datas = datas.fillna(med)
```

## 資料編碼 (One-Hot Encoding)
    1. 對字串類型欄位進行編碼
    2. 對數字編號進行編碼 (MSSubClass)

```python
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["MSSubClass"])
datas
```

## 資料分割
    將合併的資料集，拆開為
        訓練用資料集(x_train)
        預測用資料集(x_predict)
    並準備 
        訓練答案集 (y_train)
        預測用的 Id 資料集(predict_id)

```python
# 取列: .iloc -> [第一筆資料, 第二筆資料...]
x_train = datas.iloc[:train_df.shape[0]]
y_train = train_df["SalePrice"]
x_predict = datas.iloc[train_df.shape[0]:]
predict_id = test_df["Id"]
```

# 尋找合適超參

```python
from sklearn.ensemble import RandomForestRegressor
# from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import r2_score

# reg = RandomForestRegressor()
# params = {
#     'n_estimators': range(40, 80),
#     'max_depth': range(7, 14)
# }

# grid = GridSearchCV(reg, params, scoring="r2", cv=10, n_jobs=4)

# grid.fit(x_train, y_train)
# print('最佳參數: ', grid.best_params_)
# print('最佳分數 (10 次分數): ', grid.best_score_)
```

# 建立模型

```python
from sklearn.model_selection import cross_val_score

reg = RandomForestRegressor(n_estimators=56, max_depth=12)

scores = cross_val_score(reg, x_train, y_train, cv=10, n_jobs=4)

print('十次分數: ', scores)
print('平均分數: ', sum(scores) / len(scores))
```

# 訓練模型

```python
reg.fit(x_train, y_train)
```

```python
imp = reg.feature_importances_

d = sorted(zip(x_train.columns, imp),
           key=lambda x: x[1],
           reverse=True)

tables = pd.DataFrame({"Name": [n for n, im in d],
                       "Importance": [im for n, im in d]})
print(len(tables))
tables.head(20)
```

# 預測資料
    利用訓練好的模型 (reg)，對預測資料集 (x_predict) 進行預測房價 (SalePrice)

```python
pre = reg.predict(x_predict)
```

# 提交 Submit
    提交前將資料儲存成 csv，格式需符合 Kaggle 指定格式

```python
result = pd.DataFrame({
    "Id":predict_id,
    "SalePrice":pre
})
result.to_csv("house_price_rf.csv", encoding="utf-8", index=False)
result
```

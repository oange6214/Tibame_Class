# Skwe 正態分布圖
    計算數據集偏度
    
    對於正態分佈的數據，偏度應為 0 左右。偏度值 > 0表示分佈的左尾有更多權重。從統計上講，函數 skewtest 可用於確定偏度值是否足夠接近 0。

[官方](https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.skew.html)

[常態分佈密度函數教學](https://www.youtube.com/watch?v=_llGWVq5-_M)

[np.random.normal](https://blog.csdn.net/lanchunhui/article/details/50163669)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datas = np.random.normal(loc=10, scale=2, size=500000)
sns.distplot(datas) # 可以列印 長條圖、常態分布密度
```

## skew:
    計算樣本的偏數
[skew](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.skew.html)

## 資料偏度偏大

```python
from scipy.stats import skew

datas_2 = np.power(datas, 3)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```

## 資料偏度過大

```python
datas_2 = np.power(2, datas)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```

## 偏度處理
    資料偏大，開平方
    資料偏度過大 (指數成長)，取 log

```python
datas_2 = np.power(2, datas)
print("skew:", skew(np.log1p(datas_2)))
sns.distplot(np.log1p(datas_2))
```

    可以觀察到，右偏問題比較大，通常 normailize 過後的值，都只處理右偏的資料

## 左偏

```python
datas_2 = np.power(datas, 1/5)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```

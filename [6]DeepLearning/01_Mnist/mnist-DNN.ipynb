# 資料載入

```python
from tensorflow.keras.datasets.mnist import load_data

(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# 取第一筆
import pandas as pd

pd.DataFrame(x_train[0])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

print("這個是:", y_train[0])
plt.imshow(x_train[0], cmap="gray")
```

# 模型建立

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
# 64 * 784 + 64 = 50240
model.add(Dense(256, activation="relu", input_dim=784))
# 64 * 10 + 10(bias)
model.add(Dense(10, activation="softmax"))
model.summary()
```

# 模型編譯

```python
from tensorflow.keras.losses import CategoricalCrossentropy

model.compile(loss=CategoricalCrossentropy(),
              optimizer="adam",
              metrics=["accuracy"])
```

# 資料預處理 One-Hot encoding

```python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

print("原本:", y_train[2])
print("處理過:", y_train_cat[2])
```

# 正規化
    -1: 自動計算(一個位置)
    圖片: 1D -> /255(normalize)

```python
x_train_norm = x_train.reshape(-1, 784) / 255
x_test_norm = x_test.reshape(-1, 784) / 255
```

# 模型訓練
    batch_size       : 看多少筆資料做一次梯度下降(圖片不大: 100-200, 圖片大: 10-20)
    validation_split : 切出一部分作為驗證資料
    epochs           : 整份資料集看多少遍。一個 epoch 做了 54000 / 200 = 270 次梯度下降
    verbose(選用)    :  0 (Quiet), 1 (Noisy), 2 (Normal)

```python
model.fit(x_train_norm,
          y_train_cat,
          batch_size=200,
          validation_split=0.1, 
          epochs=12,
          verbose=2)
```

# 模型評估
    只要還在 tf 裡, 請帶入 one-hot encoding (模型框架性質關係)

```python
model.evaluate(x_test_norm, y_test_cat, verbose=2)
```

# 模型預測分類

```python
pre = model.predict_classes(x_test_norm)
pre
```

# 視覺化 - 混淆矩陣
    不是在 tf 裡, 帶入原本的 y

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(y_test, pre)

c = ["{}(預測)".format(i) for i in range(10)]
i = ["{}(真實)".format(i) for i in range(10)]

pd.DataFrame(mat, columns=c, index=i)
```

# 觀察 正確與錯誤答案
    nonzero: 取出是 True 的 index

```python
import numpy as np

# (x座標, y座標, z座標)
idx = np.nonzero(pre != y_test)[0]

# 預測錯誤的圖
false_img = x_test[idx]
false_answer = y_test[idx]
false_predict = pre[idx]

# 把圖變大一點才塞的下
plt.figure(figsize=(14, 42))
width = 10
height = len(false_img) // width + 1
for i, img in enumerate(false_img):
    plt.subplot(height, width, i+1)
    plt.axis("off")
    t = "[Ori]:{}\n[Pre]:{}".format(false_answer[i],
                                    false_predict[i])
    plt.title(t)
    plt.imshow(img, cmap="gray")
```

# 實際預測
    1. 載入預測圖片、轉
    2. 圖片展開、並正規化 0 ~ 1
    3. 模型預測
    4. 顯示答案、顯示影像 (轉灰階)

```python
from tensorflow.keras.preprocessing.image import load_img

fn = input("輸入檔名:")
img = load_img(fn + '.png', target_size=(28, 28)).convert("L")

# 剛剛帶入 (10000, 784)
# 現在帶入 (1, 784)
img_norm = np.array(img).reshape(1, 784) / 255
ans = model.predict_classes(img_norm)[0]

print("應該是:", ans)
plt.imshow(img, cmap="gray")
```
